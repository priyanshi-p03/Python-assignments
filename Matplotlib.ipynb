### SUBPLOT

```python
sales = np.array([36,78,67,34])
profit = np.array([34,88,77,29])
```

```python
plt.figure(figsize=(6,3))
plt.plot(sales)
plt.plot(profit)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.plot(sales, label='SALES')
plt.plot(profit, label='PROFIT')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.subplot(2,1,1)
plt.plot(profit)
plt.ylabel("Profit")
plt.subplot(2,1,2)
plt.plot(sales)
plt.ylabel("Sales")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales=np.array([27889,39867,70008,29049,60897,59978])
months=np.array(['Jan', 'Feb', 'Mar','April','May', 'Jun'])
```

```python
plt.figure(figsize=(6,3))
plt.bar(months,sales)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.barh(months,sales)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(months,sales, color='green')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales, labels=months)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales, labels=sales)
plt.show()
```

```python
label = []
for i in range(0,len(months)):
    label.append(str(months[i])+" "+str(sales[i]))
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales, labels=months)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales, labels=label)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales, labels=months, autopct="%.2f%%")
plt.show()
```
