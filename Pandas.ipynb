```python
# Creating a Dummy Dataset
data={
    'Emp_ID':[*range(101,111)],
    'Emp_Name':['Ansh','Ayush','Sneha','Parv',np.nan,'Daksh','Nishi','Vanshi','Ajay','Aryan'],
    'Emp_Salary':[35900,25680,30768,np.nan,24657,25777,34678,34575,25578,31457],
    'Emp_Add':['Noida','Gurgaon','Ghaziabad','Delhi','Pune',np.nan,'Bangalore','Gujrat','Noida','Haryana']
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
# Finding Null values in dataset
df.isnull()
```

```python
# Total null values by column
df.isnull().sum()
```

```python
# Total null values in a dataset
df.isnull().sum().sum()
```

```python
# Shape of the dataset (rows,columns)
print("Shape:",df.shape)
print("Total Rows:",df.shape[0])
print("Total Data Points:",df.shape[0]*df.shape[1])
```

```python
# Missing %age of data in dataset
print("Missing Data %age:",df.isnull().sum().sum()/(df.shape[1]*df.shape[0])*100,"%")
```
